```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', {'class': 'wikitable'})
data = []
for row in table.find_all('tr')[1:]:  
    cols = row.find_all('td')
    cols = [col.get_text(strip=True) for col in cols]
    if cols:
        data.append(cols)

df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
data = {
    'A': [10, 11, 12, 13, 14, 15, 100, 16, 17, 18, 19], 
    'B': [20, 21, 22, 23, 24, 25, 30, 26, 27, 28, 29]   
}
df = pd.DataFrame(data)
z_scores = np.abs(stats.zscore(df))
threshold = 3
outliers = (z_scores > threshold).any(axis=1)
cleaned_df = df[~outliers]
print("Original DataFrame:")
print(df)
print("\nCleaned DataFrame:")
print(cleaned_df)
```

```python
import pandas as pd
import numpy as np
data = {

    'Stock': ['AAPL', 'GOOG', 'TSLA', 'AMZN', 'MSFT'],
    'Price': [150, np.nan, 750, 3300, np.nan],
    'Volume': [10000, 15000, np.nan, 20000, 25000],
    'Change (%)': [1.5, -2.3, np.nan, 0.5, -1.2]
}
df = pd.DataFrame(data)

print("Original DataFrame with Missing Values:")
print(df)
df_dropped = df.dropna()
print("\nDataFrame after Dropping Rows with Missing Values:")
print(df_dropped)
df_filled_with_zero = df.fillna(0)
print("\nDataFrame after Filling Missing Values with 0:")
print(df_filled_with_zero)
df_filled_with_mean = df.copy()
df_filled_with_mean['Price'] = df['Price'].fillna(df['Price'].mean())
df_filled_with_mean['Volume'] = df['Volume'].fillna(df['Volume'].mean())
df_filled_with_mean['Change (%)'] = df['Change (%)'].fillna(df['Change (%)'].mean())
print("\nDataFrame after Filling Missing Values with Mean:")
print(df_filled_with_mean)
```

```python
import requests
import pandas as pd
url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    "vs_currency": "usd",  # Currency to compare against
    "order": "market_cap_desc",  # Sort by market cap descending
    "per_page": 10,  # Number of coins to fetch
    "page": 1,  # Page number for pagination
    "sparkline": "false"  # Exclude sparkline data for simplicity
}
# Make a GET request to the CoinGecko API
response = requests.get(url, params=params)
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    print(df.head())
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
import pandas as pd

import numpy as np

data = {

    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],

    'Temperature (C)': [25.3, np.nan, 18.7, 29.1, np.nan],

    'Humidity (%)': [60, 65, np.nan, 70, 55],

    'Precipitation (mm)': [2.5, 0.0, np.nan, 1.2, np.nan]

}

df = pd.DataFrame(data)

print("Original DataFrame with Missing Values:")

print(df)

df_dropped = df.dropna()

print("\nDataFrame after Dropping Rows with Missing Values:")

print(df_dropped)

df_filled_with_value = df.fillna(-1)

print("\nDataFrame after Filling Missing Values with -1:")

print(df_filled_with_value)

df_filled_with_mean = df.copy()  

df_filled_with_mean['Temperature (C)'] = df['Temperature (C)'].fillna(df['Temperature (C)'].mean())

df_filled_with_mean['Humidity (%)'] = df['Humidity (%)'].fillna(df['Humidity (%)'].mean())

df_filled_with_mean['Precipitation (mm)'] = df['Precipitation (mm)'].fillna(df['Precipitation (mm)'].mean())

print("\nDataFrame after Filling Missing Values with Mean:")

print(df_filled_with_mean)
```
